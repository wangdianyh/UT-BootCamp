```python
import pandas as pd
import time
from datetime import datetime
```

```python
path = 'employee_data.csv'
df = pd.read_csv(path)
df.head()
```

```python
export = pd.DataFrame()
export['Emp'] = df['Emp ID']
export.head()
```

```python
#export = pd.DataFrame(df['Name'].str.replace(' ', ','), columns=['Name'])
name = df['Name'].str.split(" ", n=1, expand = True)
export['First Name'] = name[0]
export['Last Name'] = name[1]
export.head()
```

```python
dob = []
for row in df['DOB']:
    timeS = time.strptime(row, '%Y-%m-%d')
    timef = time.strftime('%m/%d/%Y', timeS)   
    dob.append(timef)

export['DOB'] = pd.Series(dob)
export.head()
```

```python
SSN = []
for row in df['SSN']:
    new_row = row.replace(row[0:6], '***-**')
    SSN.append(new_row)
    
export['SSN'] = pd.Series(SSN)
export.head()
```

```python
abb = df['State'].unique()
abb = sorted(abb)
state_map = pd.DataFrame(data=abb, columns=['state'], index=['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA',
                                                          'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 
                                                          'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ',
                                                          'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC',
                                                          'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY'])
state_map.head()
```

```python
state_label = []
for row in df['State']:
    add_row = state_map.index[row == state_map['state']].item()
    state_label.append(add_row)
    
export['State'] = pd.Series(state_label)
export.head()
```

```python
export.to_csv('pyboss_export.csv', encoding='utf-8', index=False, header=True)
```
